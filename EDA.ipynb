# Análise Bivariada

```python
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, margins_name='Total')
```

```python
pd.crosstab(df_churn.Churn, df_churn.Contract, normalize='index', margins=True, margins_name='Total')
```

Considerando o público total, os cliente de contrato mensal representam 55%. Porém, se considerarmos apenas os cliente que abandonaram o serviço, 88% tinham contrato mensal. Desta forma, podemos afirmar que há uma correlação entre o tipo de contrato mensal e o abandono.

```python
# Avaliar a correlação entre duas variáveis categóricas (qualitativas)

# Executar teste de hipótese 
# H0 (Hipótese nula): as duas variáveis são independentes
# H1 (Hipótese complementar): as duas variáveis não são independentes

# O teste é usado para confirmar ou recusar a hipótese nula (H0)
# Quando a propabilidade de observarmos H0 é inferior a 0.05 (p-value), recusamos a hipótese nula e seguimos com a complementar
```

```python
df_crosstab_churn_contract = pd.crosstab(df_churn.Churn, df_churn.Contract)
```

```python
from scipy.stats import chi2_contingency
```

```python
chi_scores_churn_contract = chi2_contingency(df_crosstab_churn_contract)
```

```python
chi_scores_churn_contract
```

```python
scores_churn_contract = pd.Series(chi_scores_churn_contract[0])
```

```python
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
```

```python
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
df_chi_scores_churn_contract = pd.DataFrame({'Qui2': scores_churn_contract, 'p-value': pvalues_churn_contract})
```

```python
df_chi_scores_churn_contract
```

```python
# P-Value é menor que 0.05, portanto descartamos a hipótese nula (H0). Concluímos que as variáveis não são independentes
```

```python
# Calcular scores
chi_scores_churn_tenure = chi2_contingency(df_crosstab_churn_tenure)
```

```python
# Criar séries
scores_churn_tenure = pd.Series(chi_scores_churn_tenure[0])
pvalue_churn_tenure = pd.Series(chi_scores_churn_tenure[1])
```

```python
# Cria um DataFrame com as séries criadas acima
df_chi_scores_churn_tenure = pd.DataFrame({'Qui2': scores_churn_tenure, 'p-value': pvalue_churn_tenure})
```

```python
df_chi_scores_churn_tenure
```

```python
# P-Value < 0,05 portanto as variáveis não são independentes
# Vale mencionar que com base no Qui2 resultante das 2 análises, a correlação entre Churn e Tempo de Contrato < 6 meses é menor do que a correlação entre Churn e Tipo de Contrato
```

```python
df_churn_month
```

```python
df_churn_month.TotalCharges.plot.hist()
```

```python
# Método de Turkey - IQR (Distribuição não normal)
# IQR: Range interquartil
# IQR: Terceiro quartil - Primeiro quartil
q1_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.25)
q3_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.75)
iqr_TotalCharges_month = q3_TotalCharges_month - q1_TotalCharges_month
iqr_TotalCharges_month
```

```python
# Limites Inferior e Superior
limInf_TotalCharges_month = q1_TotalCharges_month - (iqr_TotalCharges_month * 1.5)
limSup_TotalCharges_month = q3_TotalCharges_month + (iqr_TotalCharges_month * 1.5)
limInf_TotalCharges_month, limSup_TotalCharges_month
```

```python
df_churn_month[(df_churn_month.TotalCharges < limInf_TotalCharges_month) | (df_churn_month.TotalCharges > limSup_TotalCharges_month)]
```

```python
# Técnica ZScore (Usando para Distribuição Normal)
# Indica quantos desvios padrões um ponto específico de dados está distante da média
# zscore = (x - média) / desvio padrão
from scipy.stats import zscore
```

```python
# Função np.abs => Retorna números absolutos
z = np.abs(zscore(df_churn_month.TotalCharges))
```
